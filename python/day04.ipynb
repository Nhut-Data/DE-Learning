```python
#Error Handling & Context Managers
#truy/ except/ else/ finally
def safe_devide(a: float, b: float):
    try:
        result = a/ b
    except ZeroDivisionError: #ktr lỗi chia cho 0
        print('lỗi: ko thể chia cho 0')
        return None
    except TypeError as e: # ktr lỗi chia cho string
        print(f"ko thể chia cho {e}")
        raise
    else:
        #chạy khi có except 
        print(f"chạy thành công : {result}")
        return result
    finally:
        #luôn chạy bất kể có lỗi hay ko
        print('chạy ngay đây')

print(safe_devide(2,1))
```

```python
#custom xceptions
class DataValidationError(ValueError): #gọi class với class cha lỗi liểu dữ liệu
    '''raise khi data ko hợp lệ'''
    def __init__(self, field: str, message: str): # tạo hàm khởi tạo
        self.field = field
        self.message = message
        super().__init__(f"validation error on {field}: {message}") #gọi lớp cha của class hiện tại là ValueError
        
class PipelineError(RuntimeError): # lớp cha với runtimeerror
    '''raised khi pipeline fail'''
    pass

#dùng:
def validate_email(email: str) -> str: # email nhập vào str và trả về cx là str
    if "@" not in email: # nếu ko có @ tỏng email thì 
        raise DataValidationError("email", f"lỗi dữ liệu ở {email}") # trỏ tới class để báo lỗi với datavalidationerror 
    return email.lower().strip()

try: 
    validate_email('minhnhut@123')
except DataValidationError as e:
    print(f"field: {e.field}, lỗi: {e.message}")
```

```python
#context managewrs - with statement
# tự độnhg cleanup dù có exception hay ko
 #file handling
with open('data.csv', 'r', encoding = 'utf-8') as f:
    content = f.read()

with open('input.csv') as fin, open('output.csv', 'w') as fout:
    fout.write(fin.read().upper())
    
#custom context manager với contextlib

from contextlib import contextmanager
import time

@contextmanager
def timer(name: str):
    '''đo thời gian được cảu mỗi block code'''
    start = time.perf_counter()
    try:
        yield # code trong with block chạy ở đây
    finally:
        elasped = time.perf_counter() - start
        print(f'{name} {elasped:.3f}s')
def run_extract():
    print("extracting...")

def run_transform():
    print("transforming...")

def run_load():
    print("loading...")
    
with timer('ETL Pipeline'):
    run_extract()
    run_transform()
    run_load()

    
```
