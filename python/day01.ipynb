```python
#list - tuple - set - dict
#list
nums = [1,2,3,5,6,8,9]
nums.append(7) #thêm phần tử vào cuối danh sách
nums.insert(-1, 4) #thêm hần vào index
nums.remove(9) #xóa phần tử có giá trị 9
nums.extend([10,22,33]) #thêm nhiều phần tử vào cuối danh sách
val1 = nums.pop() #xóa và trả về phần tử cuối
val2 = nums.pop(0) #xóa và trả về phần tử index 0
nums.sort() # sắp xếp danh sách
nums.sort(reverse = True) #sắp xếp danh sách theo thứ tự giảm dần
sorted_nums = sorted(nums) #trả về một danh sách mới đc sắp xếp
nums.reverse() #đảo ngược danh sách
print(nums.index(5)) #in ra index có giá trị 5
print(nums.count(2)) # đém số lần xuất hiện của 2
print(len(nums)) # đếm số phần tử của danh sách
print(nums)
#list clicing
first_3 = nums[:3] #lấy 3 phân tử dầu tiên
last_3 = nums[-3:] # lấy 3 phần tử cuối
copy = nums[:] # sao chép danh sách
every_2 = nums[::2] # lấy mỗi phần tử thứ 2

```

```python
#set
emails = {'hehe@gmail.com', 'haha@gmail.com', 'hihi@gmail.com'}
print(emails)

a = {1,2,3,4}
b = {6,3,4,5}
print(a | b) # hợp 2 set 
print(a & b) # giao 
print(a ^ b) #loại bỏ phần tủ chung
print(a - b) #loại bỏ phần tử a khỏi b

#dùng set để dedup nhanh
raw_id = [1,2,2,3,4,6,5]
unipue = list(set(raw_id)) # dedup
print(unipue)

```

```python
#dict
customer = {
    'id': 1,
    'name': 'nhựt',
    'age': 20,
    'email': 'nhutcute@gamil.com',
    'tier': 'gold'
}

#truy cập
print(customer['name']) # keyerror nếu ko có
print(customer.get('phone')) # None - ko raise error
print(customer.get('phone', 'N/A'))

#thêm sửa xóa
customer['name'] = 'dương' # thêm or update
customer.update({'tier': 'platium', 'age': 19}) # update
del customer['id'] # keyerror nếu ko có
tier = customer.pop('tier', None) # xóa và trả về value

#Iteration
for key in customer: # duyệt các key trong customer
    print(key)
for key, value in customer.items():
    print(f'key: {key}, valie: {value}')
key = list(customer.keys()) #
value = list(customer.values())

```

```python
# Dict methods hay dùng trong DE
# setdefault — set nếu key chưa tồn tại

count = {}
for city in ['hcm', 'dn', 'hn']:
    count.setdefault(city, 0) # đếm và tạo value
    count[city] += 1
print(count)
#cách viết khác
huhu = {}
for city in ['haha', 'hihi', 'hehe']:
    huhu[city] = huhu.get(city, 0 ) + 1
print(huhu)
#đếm dùng thư viện
from collections import Counter

icon = ['hihi', 'hehe', 'huhu']
print(Counter(icon))

#defaultdict - tự động tạo default value
from collections import defaultdict
counts = defaultdict(int) #default value = 0
for city in ['hcm', 'hcm', 'dl', 'hue']:
    counts[city] += 1 # không cần setdefault
print(counts)

```
