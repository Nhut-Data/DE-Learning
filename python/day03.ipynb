```python
# List, Dict, Set Comprehensions
# List Comprehension
# [expression for item in iterable if condition]
#cách viết bình thường 
squares = []
for x in range(10):
    squares.append(x ** 2) # chạy các x từ 0 tói 9 và thêkm vào list với môi x mũ 2
print(squares)

#cáh gọn hơn 
square = [x**2 for x in range(10)] 
print(square)

# với condition
squared = [x**2 for x in range(20) if x % 2 == 0] # chạy các x từ 0 tới 20 nếu x đó là số chãn và thêm vào list với x mũ 2
print(squared)
 
#ứng dụng trong DE
rows = [{
    'id': 1,
    'amount': 100
}, {
    'id': 2,
    'amount': 400
}, {
    'id': 3,
    'amount': 1000
}]

valid_amounts = [va['amount'] for va in  rows if va['amount'] > 200] # duyệt qua các amount có trên 200
order_id = [oi['id'] for oi in rows] # hiện id của các dict 
clean_row = [cr for cr in rows if cr['amount'] is not None] # in ra các thông tin của rows nếu các amount cảu các dict đều khác none
print(valid_amounts)
print(order_id)
print(clean_row)
```

```python
# Nested comprehension — transform data
data = [['1   ', 'nhựt   ', '  hcm'], ['2  ', 'dương  ', '  nghean']]
clean = [[cell.strip() for cell in row] for row in data] # làm sạch dữ liệu xóa khoản trắng bằng strip và duệt qua dòng và data
print(clean)

#Dict Comprehension 
names = ['nhựt', 'dương', 'linh']
scores = [10, 0, 9]

# zip -> dict
grade_map = {name: score for name, score in zip(names, scores)} # chuyển đổi 2 list thành 1 dict bằng comprehension 
print(grade_map)

#invert dict 
id_to_name = {
    1: 'nhựt',
    2: 'dương',
    3: 'dương'
}
name_to_id = {name: id for id, name in id_to_name.items()} # đổi chổ key và value bằng cách đổi chổ biến for
print(name_to_id)
```

```python
# Transform values
revenue = {'electronics': 500_000_000,
           'mobile': 10_000_000}
revenue_m = {cat: val/1_000_000 for cat , val in revenue.items()} # duệt qua các cat và val trong bảng cách chia chia value 
print(revenue_m)

# Filter + transform
big_revenue = {cat: val for cat, val in revenue.items() if val > 40_000_000} #tương tụ câu trên
print(big_revenue)

```

```python
#FUNCTIONS
# function basic
def greet(name, greeting):
    return f'{greeting}, {name}'

greet('nhựt', 'hello') # quá cơ bản rồi

#keyword agrs - rõ ràngm thứ tự ko quan trọng
greet(name = 'dương', greeting = 'hi') 

#*agrs và **kwagrs
# với args là nhận nhiều giá trị tùy ý còn kwargs nhận nhiều key = value tùy ý
def log(*args, **kwagrs): 
    timetamp = kwagrs.get('timetamp', 'now')
    level = kwagrs.get('level', 'info')
    message = ' ' .join(str(a) for a in args)
    print(f"[{timetamp}] [ {level}] {message}")

log('ETL', 'pipeline', level = 'info', timetamp = '2024-01-15')

# unpack khi gọi function
def add(a,b,c):
    return a+b+c

hehe = [1,2,3]
add(*hehe)

config = {'a': 1, 'b': 3, 'c': 5}
add(**config)

```

```python
#tyoe hints
from typing import Optional, Union, Any
from collections.abc import Callable , Iterator, Generator

def process_records(
    records: list[dict[str: Any]],
    filter_fn: Callable[[dict], bool] | None = None,
    batch_size: int = 100,
) -> list[dict[str, Any]]:
        '''
    Process records với optional filter function.
    Args:
        records: list of record dicts
        filter_fn: optional function to filter records
        batch_size: number of records per batch
    Returns:
        Processed records
    '''
        if filter_fn:
            records = [r for r in records if filter_fn(r)]
        return records

```
