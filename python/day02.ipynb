```python
#day 02 for python
#control flow
#if / elif / else

score = float(input('nhập số điểm: '))
if score > 90 :
    grade = 'A'
elif score > 60:
    grade = 'B'
else:
    grade = 'C'
print(grade)

#ternary expression
status = 'pass' if score > 60 else 'fail'
print(status)
```

```python
#match, case (ngon hơn if/else)
def process_status(status: str):
    match status:
        case 'completed': return 'seccess'
        case 'minhnhut': return 'td bi khùng'
        case 'hehe': return 'hihi'
        case 'fails' | 'error': return 'needs retry'
        case _: return 'unknow'

process_status('hehe')   
 
```

```python
#traing match/case
#câu 1 basic
day = int(input('nhập số từ 1-7: '))
match day:
    case 1: print('thứ 2')
    case 2: print('thứ 3')
    case 3: print('thứ 4')
    case 4: print('thứ 5')
    case 5: print('thứ 6')
    case 6: print('thứ 7')
    case 7: print('chủ nhật')
    case _: print('error, mời nhập lại trong khoảng 1->7')
    
#câu 2
nums = int(input('nhập số: '))
match nums:
    case 2 | 4 | 6: print('số chẵn')
    case nums if nums > 10: print('số lớn')
    case _: print('bình thường')

#câu 3 | vì câu 3 này bro ko quy định rõ số nguyên hay thực và có nhập string ko nên tui quy về số thực và ko có string 
lists = list(map(float, input('nhập số vào list: ').split()))
print(lists)
so_phan_tu = len(lists)
match so_phan_tu:
    case so_phan_tu if so_phan_tu == 2: print(f'tổng = {lists[0] + lists[1]}')
    case so_phan_tu if so_phan_tu >= 3: print(f'first = {lists[0]}, rest = {lists[1:]}')
    
#câu 4
data = {
    'type': 'student',
    'name': 'nhut',
    'socre': 8
}
match data['type']:
    case 'student': print(f'{data['name']}, {data['score']}')
    case 'teacher': print('giáo viên')
    case _: print('lỗi dữ liệu')

#câu 5 | AI viết vì nó lỡ làm luôn rồi
data = ["move", (-2, 10)]
match data:
    case ["move", (0, 0)]:
        print("Đứng yên")
    case ["move", (x, _)] if x > 0:
        print("Đi sang phải")
    case ["move", (x, _)] if x < 0:
        print("Đi sang trái")
    case ["stop"]:
        print("Dừng lại")
    case _:
        print("Không hợp lệ")
    
```

```python
#loops
#for loop
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit, end=", ")

#enumerate - lấy cả index và value
for i, e in enumerate(fruits):
    print(f"index: {i}, value: {e}")

for i, e in enumerate(fruits, start = 1):
    print(f"index: {i}, vlaue: {e}")

#zip - iterate nhiều iterables cùng lúc
names = ['nhựt', 'dương']
ages = [20, 19]
for n, a in zip(names, ages):
    print(f"tên: {n}, có tuổi: {a}")

#zip_longest - ko dừng khi hết list ngắn hơn
from itertools import zip_longest
for a, b in zip_longest([1,2,3], [20,30], fillvalue = 'ko có value'):
    print(a, b)

# range - quá cơ bản rồi nên copy paste cho lẹ hihi
for i in range(10):    # lấy từ 0 - 9  
    print(i)
for i in range(1, 11):     #lấy từ 1 - 10
    print(i)
for i in range(0, 100, 10): # lấy từ 0 - 99 cách nhau 10 đơn vị
    print(i)

#while loop
def fetch_data():
    return ConnectionError("lỗi mạng nè con")
retry = 0
while retry < 3:
    try:
        result = fetch_data()
        print(f"thành công : {result}")
        break                   #thoát vòng lâp thành công
    except ConnectionError:
        retry += 1
        print(f"retry: {retry} / 3")
else:
    print('all retries fails ')  # chạy khi while kết thúc ko phải bởi break
    
# break , continue, pass
for row in data:
    if row is None: continue  # bỏ qua row này
    if row["erroe"]: break  # dùng vòng lặp
    pass # 
```
